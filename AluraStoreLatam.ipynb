#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
#Verificamos que no haya nulos en el dataset
tienda3.isnull().sum()
```

```python
#Despues de confirmar que no hay nulos en el dataset, asignaremos variables a la suma de ventas de cada tienda
venta_total_tienda1=tienda.Precio.sum()
venta_total_tienda2=tienda2.Precio.sum()
venta_total_tienda3=tienda3.Precio.sum()
venta_total_tienda4=tienda4.Precio.sum()
#Mostramos la venta total de cada tienda
print(f"La venta total de la tienda 1 es {venta_total_tienda1}")
print(f"La venta total de la tienda 2 es {venta_total_tienda2}")
print(f"La venta total de la tienda 3 es {venta_total_tienda3}")
print(f"La venta total de la tienda 4 es {venta_total_tienda4}")
```

```python
#Agregaremos un grafico para mejor visualizacion
import matplotlib.pyplot as plt

# Creamos un dataframe con las columnas Tienda y Venta Total, invocaremos directamente la suma de cada tienda para evitar errores
df_ventas = pd.DataFrame(
    {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Venta Total': [tienda['Precio'].sum(),tienda2['Precio'].sum(),tienda3['Precio'].sum(),tienda4['Precio'].sum()]
})
# Graficamos asignando los valores, porcentaje , colores etc
plt.pie(
    df_ventas['Venta Total'],# Valores
    labels=df_ventas['Tienda'],# Etiquetas
    autopct='%1.1f%%',# Mostrar porcentaje con un decimal
    colors=['skyblue', 'lightgreen', 'orange', 'yellow'],  # Colores
)
plt.title('Distribución de Ventas Totales por Tienda', fontsize=14)
plt.show()
```

##Observacion
Decidi utilizar un grafico tipo torta o pie, para poder ver el porcentaje que aporta cada tienda respecto a las ventas totales
De esta forma podemos observar que la tienda que mas vendio es la Tienda 1 y la que menos vendio es la Tienda 4

## Análisis geográfico de las ventas

Como parte del análisis adicional, se utilizaron los datos de **latitud y longitud** de las transacciones para representar visualmente la distribución de las ventas en un mapa.  
El objetivo fue identificar las zonas con **mayor concentración o monto total de ventas**, utilizando un **mapa de calor (HeatMap)** generado con la librería *Folium*.

A continuacion el mapa:

```python
!pip -q install folium
import folium
from folium.plugins import HeatMap

# Combinamos los datos de totas las tiendas de forma que los datos de las 4 tiendas se hacen una
df_all = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
#Como buenas pracitcas borramos todos los nulos, si es que hubieran
df_all = df_all.dropna(subset=['lat','lon'])

# Creamos una variable center para que el mapa se centre de acuerdo al promedio de la longitud y latitud de los datos
center = [df_all['lat'].mean(), df_all['lon'].mean()]

# Creamos un mapa base con folium la locacion principal sera el center
m = folium.Map(location=center, zoom_start=5, tiles='OpenStreetMap')

# Creamos un heat data con los datos latitud, longitud y el precio, los datos de latitud y longitud seran para ubicar en el mapa el punto, y el Precio sera para
# la intensidad del punto
heat_data = df_all[['lat','lon','Precio']].values.tolist()
HeatMap(heat_data, radius=10, blur=15, max_zoom=12).add_to(m)
#Creamos el mapa y lo abrimos
m
```

##Conclusion EXTRA
En el mapa se observa que **Bogotá** presenta la zona más brillante, lo que indica una **mayor densidad de ventas o montos más altos** en comparación con las demás regiones, esto probablemente se debe a que es la capital de Colombia.  
Este resultado sugiere que la capital concentra un volumen más significativo de transacciones, posiblemente por su **alta actividad comercial, mayor número de clientes potenciales y capacidad de compra**.

Las demás zonas muestran niveles similares de actividad, con menor intensidad en el mapa, lo que evidencia que las ventas están más dispersas o son de menor valor unitario.  
Por tanto, **Bogotá representa el principal foco de ingresos**, siendo el punto estratégico con mejor rendimiento en el análisis geográfico de ventas.
